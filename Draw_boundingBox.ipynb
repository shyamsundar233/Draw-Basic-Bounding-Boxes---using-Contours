```python
import os
import cv2
from PIL import Image, ImageEnhance
from google.colab.patches import cv2_imshow
import argparse
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def bbox(path):
  img = cv2.imread(path)
  img-=10
  imgray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
  org = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
  
  '''
  lower = np.array([30,30,30])
  higher = np.array([250,250,250])

  mask = cv2.inRange(img,lower,higher)
  '''

  ret,thresh = cv2.threshold(imgray,127,255,0)
  contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)
  cont_img = cv2.drawContours(img,contours,-1,(0,255,0),3)
  #plt.imshow(cont_img)
  #plt.show()

  c = max (contours, key= cv2.contourArea)

  org = cv2.imread(path)
  org = cv2.cvtColor(org,cv2.COLOR_BGR2RGB)
  x,y,w,h = cv2.boundingRect(c)
  org = cv2.rectangle(org,(x,y),(x+w,y+h),(0,255,0),5)
  plt.imshow(org)
  plt.show()
```
